# Reflectivity

We define the Reflectivity of a system as:

$$ R(\omega)=\frac{(n-1)^2+k^2}{(n+1)^2+k^2} $$

where:

$\omega$ is the given frequency;

$n$ is refractive index;

$k$ is extinction coefficient.

References:

    vaspkit.com/tutorials.html#linear-optical-properties

```python
# Combine 3 raster figures

from PIL import Image
import matplotlib.pyplot as plt

from vmatplot.output import canvas_setting, color_sampling

def merge_3_rasters(image_paths):
    # Figure settings
    fig_setting = canvas_setting()
    annotate_color = color_sampling("Grey")

    # Load images
    images = [Image.open(image) for image in image_paths]

    # Assuming all images have the same size
    width, height = images[0].size

    # Create a new image to merge the four images
    result = Image.new("RGB", (width, 3 * height))

    # Paste the images into the new image
    result.paste(images[0], (0, 0))
    result.paste(images[1], (0, height))
    result.paste(images[2], (0, 2*height))

    # Create a matplotlib figure
    # _, ax = plt.subplots()
    _, ax = plt.subplots(figsize=(fig_setting[0][1],fig_setting[0][0]), dpi=fig_setting[1])
    params = fig_setting[2]; plt.rcParams.update(params)
    ax.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)
    ax.imshow(result)

    # Set ticks at the junctions of the images
    plt.xticks([width - 0.5], [""])  # Horizontal junction
    plt.yticks([height - 0.5, 2 * height - 0.5], ["",""])  # Vertical junctions

    # Dividing lines
    line_color = "black"
    for i in range(1, 3):
        plt.axhline(y=i * height, color=line_color, linewidth=1)

    # Add labels to each image
    labels = ["a","b","c"]
    for i, label in enumerate(labels):
        x = 0.075 * width
        y = (3 - i) * height - 0.100 * height
        ax.annotate(f"({label})",
                    xy=(x / (width), y / (3 * height)),
                    xycoords="axes fraction",
                    fontsize=1.0 * 16,
                    ha='center', va='center',
                    bbox = {"facecolor": "white", "alpha": 0.75, "edgecolor": annotate_color[2], "linewidth": 1.5, "boxstyle": "round, pad=0.2"}, zorder=5)

    plt.show()

biimg_path = ["Supercells/Ideal/E_Graphene-BC3_Hollow.png",
                "Supercells/Ideal/F_Graphene-Borophene_Top.png",
                "Supercells/Ideal/G_Graphene-B4C3_Top.png"]

merge_3_rasters(biimg_path)
```

```python
# Combine 4 raster figures

from PIL import Image
import matplotlib.pyplot as plt

from vmatplot.output import canvas_setting, color_sampling

def merge_4_rasters(image_paths):
    # Figure settings
    fig_setting = canvas_setting()
    annotate_color = color_sampling("Grey")

    # Load images
    images = [Image.open(image) for image in image_paths]

    # Assuming all images have the same size
    width, height = images[0].size

    # Create a new image to merge the four images
    result = Image.new("RGB", (2 * width, 2 * height))

    # Paste the images into the new image
    result.paste(images[0], (0, 0))
    result.paste(images[1], (width, 0))
    result.paste(images[2], (0, height))
    result.paste(images[3], (width, height))

    # Create a matplotlib figure
    _, ax = plt.subplots(figsize=fig_setting[0], dpi=fig_setting[1])
    params = fig_setting[2]; plt.rcParams.update(params)
    ax.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)
    ax.imshow(result)

    # Set ticks at the junctions of the images
    plt.xticks([width - 0.5],  [""])  # Horizontal junction
    plt.yticks([height - 0.5], [""])  # Vertical junction

    # Dividing lines
    line_color = "black"
    plt.axhline(y=1 * height, color=line_color, linewidth=1)
    plt.axvline(x=1 * width, color=line_color, linewidth=1)

    # Add labels to each image
    labels = ["a","b","c","d"]
    for i, label in enumerate(labels):
        # if i == 0:
        #     x = (1+i% 2) * width  -  0.075 * width
        #     y = (1-i//2) * height + 0.100 * height
        # elif i == 1:
        #     x = (i %  2) * width  +  0.075 * width
        #     y = (1-i//2) * height + 0.100 * height
        # elif i == 2:
        #     x = (1+i% 2) * width  -  0.075 * width
        #     y = (i // 2) * height - 0.100 * height
        if i in [0,1]:
            x = (i %  2) * width  +  0.075 * width
            y = (2+i//2) * height - 0.100 * height
        elif i in [2,3]:
            x = (i %  2) * width  +  0.075 * width
            y = (i // 2) * height - 0.100 * height
        ax.annotate(f"({label})",
                    xy=(x / (2 * width), y / (2 * height)),
                    xycoords="axes fraction",
                    fontsize=1.0 * 16,
                    ha='center', va='center',
                    bbox = {"facecolor": "white", "alpha": 0.75, "edgecolor": annotate_color[2], "linewidth": 1.5, "boxstyle": "round, pad=0.2"}, zorder=5)

    plt.show()

monoimg_path = ["Supercells/Ideal/A_BC3.png",
                "Supercells/Ideal/B_Borophene.png",
                "Supercells/Ideal/C_B4C3.png",
                "Supercells/Ideal/D_Graphene.png"]

merge_4_rasters(monoimg_path)
```

# PDoS elements

Projected electronic density of state versus energy for each elements


```python
# Prework

from vmatplot.PDoS import extract_pdos, extract_segment_pdos
from vmatplot.PDoS_plotting import plot_segment_pdos_fit

import matplotlib.pyplot as plt
```

```python
# Graphene-BC₃ (Hollow)

HSE_total = extract_pdos("4_PDoS/E_Graphene-BC3_Hollow_HSE")
HSE_B     = extract_segment_pdos("4_PDoS/E_Graphene-BC3_Hollow_HSE", 1, 2)
HSE_C_top = extract_segment_pdos("4_PDoS/E_Graphene-BC3_Hollow_HSE", 3, 8)
HSE_C_bot = extract_segment_pdos("4_PDoS/E_Graphene-BC3_Hollow_HSE", 9, 16)

title_total = "Total PDoS"
title_C_top = "PDoS for Carbon in Graphene"
title_B_BC3 = "PDoS for Boron in BC₃"
title_C_BC3 = "PDoS for Carbon in BC₃"

matters_dir = [
    [title_total, 6, 10,    "", HSE_total],
    [title_C_top, 6, 3.0,   "", HSE_C_top],
    [title_B_BC3, 6, 3.0,   "", HSE_B],
    [title_C_BC3, 6, 3.0,   "", HSE_C_bot],
    ["blue", 1.0, "solid"]
]

plot_segment_pdos_fit("for Graphene-BC₃ (HSE06)", matters_dir)
plt.savefig("Figures/4_pdos_Graphene-BC3_fit.pdf")
```

```python
# Graphene-Borophene (Top)

HSE_total = extract_pdos("4_PDoS/F_Graphene-Borophene_Top_HSE")
HSE_B     = extract_segment_pdos("4_PDoS/F_Graphene-Borophene_Top_HSE", 1, 8)
HSE_C_top = extract_segment_pdos("4_PDoS/F_Graphene-Borophene_Top_HSE", 9, 16)

title_total = "Total PDoS"
title_C_top = "PDoS for Carbon in Graphene"
title_B     = "PDoS for Boron in Borophene"

matters_dir = [
    [title_total, 6, 10,    "", HSE_total],
    [title_C_top, 6, 4.0,   "", HSE_C_top],
    [title_B,     6, 4.0,   "", HSE_B],
    ["blue", 1.0, "solid"]
]

plot_segment_pdos_fit("for Graphene-Borophene (HSE06)", matters_dir)
plt.savefig("Figures/4_pdos_Graphene-Borophene_fit.pdf")
```

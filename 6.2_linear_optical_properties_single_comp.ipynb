```python
from vmatplot.linear_optical_properties import *
import matplotlib.pyplot as plt

def plot_lop_monocomp(suptitle, systems_list=None, properties=None, component="xx", layout="horizontal",
                      unit="eV", boundary=(None,None), figure_size=(None,None)):
    ## Help information
    current_help = f"Usage: plot_lop_monocomp \n" +\
                   f"Demonstrate linear optical properties by each component \n" +\
                   "\t suptitle: the suptitle; \n" +\
                   "\t systems_list: dielectric function data list; \n" +\
                   "\t properties: select properties in a list; \n" +\
                   "\t components: select one component ({'xx'<default>, 'yy', 'zz', 'xy', 'yx', 'yz', 'zy', 'zx', 'xz'}); \n" +\
                   "\t layout: subfigures layout (horizontal<default>, vertical); \n" +\
                   "\t unit: x-axis unit (eV<default>, nm); \n" +\
                   "\t boundary: a-axis range <optional>; \n" +\
                   "\t figure_size: figure size <optional>. \n"
    if suptitle.lower() =="help":
        print(current_help)
        return

    ## components aliases
    component_label, comp_aliase = None, None
    if isinstance(component, list):
        component = component[0]
    if isinstance(component, dict):
        key, value = next(iter(component.items()))
        component_label = key.lower()
        comp_aliase = value
    else:
        component_label = component.lower()
        comp_aliase = f"{component}-component"

    ## scale flag and databoundaries
    rescale_flag, source_start, source_end, scaled_start, scaled_end = process_boundaries_scaling(boundary)

    ## properties
    multi_prop_flag = None
    if isinstance(properties, list) and len(properties)>1:
        multi_prop_flag = True
        formula_titles, formula_flags, compfunc_names = [],[],[]
        for current_prop in properties:
            current_prop_dict = identify_linear_optical_functions(current_prop)
            current_formula_title = current_prop_dict["title"]
            current_formula_flag = current_prop_dict["flag"]
            current_compfunc_name = current_prop_dict["calculation function"]
            formula_titles.append(current_formula_title)
            formula_flags.append(current_formula_flag)
            compfunc_names.append(current_compfunc_name)
    else:
        multi_prop_flag = False
        prop_dict = identify_linear_optical_functions(properties)
        formula_title = prop_dict["title"]
        formula_flag = prop_dict["flag"]
        compfunc_name = prop_dict["calculation function"]
        formula_titles = formula_title
        formula_flags = formula_flag
        compfunc_names = compfunc_name
    
    print(formula_titles, formula_flags, compfunc_names)

    ## flags and figure settings
    layout_flag = "horizontal" if layout.lower() not in ["vertical", "ver","v"] else "vertical"
    single_subfigure_flag = False
    if multi_prop_flag == False and rescale_flag == False:
        single_subfigure_flag = True
        fig_setting = canvas_setting() if figure_size == (None, None) else canvas_setting(figure_size[0], figure_size[1])
        plt.figure(figsize=fig_setting[0], dpi=fig_setting[1])
        params = fig_setting[2]
        plt.rcParams.update(params)
    # # single prop with rescale view
    # elif multi_prop_flag == False and rescale_flag == True:

    # # multi prop without rescale view
    # elif multi_prop_flag == True and rescale_flag == False:

    # # multi prop with rescale view
    # elif multi_prop_flag == True and rescale_flag == True:


    ## identify x-axis unit
    var_label = "wavelength" if unit and unit.lower() == "nm" else "energy"
    xaxis_str = "Photon wavelength (nm)" if var_label == "wavelength" else "Photon energy (eV)"


systems = [["Graphene-BC₃ (17 k-points)", "5_Dielectric_function/E_Graphene-BC3_HSE_K17", "blue"],
           ["Graphene-Boropheneby", "5_Dielectric_function/F_Graphene-Borophene_HSE_K65_Accurate", "green"],
           ["Graphene-B₄C₃ (17 k-points)", "5_Dielectric_function/G_Graphene-B4C3_HSE_K17", "violet"]]

# components = [{"xx":"in-plane"}, {"zz":"out-of-plane"}, "xy", "yx","zx"]
component = [{"xx":"in-plane"}]
boundary = ((0,40),(4,12))
boundary = 24
layout = "hor"

plot_lop_monocomp("for bilayers by HSE06", systems, "absorption", component, layout, "eV", boundary)
```

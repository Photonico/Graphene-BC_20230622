```python
from vmatplot.linear_optical_properties import *
from vmatplot.commons import process_boundaries_scaling
import matplotlib.pyplot as plt

def plot_linear_optical_property(suptitle, systems_list=None, properties=None, components="xx", systems_expansion=False,
                                 layout="horizontal", unit="eV", boundary=(None, None), figure_size=(None, None)):
    # Support information
    if suptitle.lower() == "help, support":
        help_info = lop_plotting_help()
        print(help_info)
        return

    # boundaries processing
    rescale_flag, source_start, source_end, scaled_start, scaled_end = process_boundaries_rescaling(boundary)

    # multi comp determination
    multi_comp_flag = None
    comp_labels, comp_aliases = [], []
    if isinstance(components, str):
        multi_comp_flag = False
        comp_labels.append(components)
        comp_aliases.append(f"{components}-component")
    elif isinstance(components, dict):
        if len(components) == 1:
            multi_comp_flag = False
            comp_labels.append(list(components.keys())[0])
            comp_aliases.append(list(components.values())[0])
        elif len(components) > 1:
            multi_comp_flag = True
            comp_labels.append(list(components.keys()))
            comp_aliases.append(list(components.values()))
    elif isinstance(components, list):
        if len(components) == 1:
            multi_comp_flag = False
        elif len(components) > 1:
            multi_comp_flag = True

        for comp_unit in components:
            if isinstance(comp_unit, dict):
                comp_labels.append(list(comp_unit.keys())[0])
                comp_aliases.append(list(comp_unit.values())[0])
            elif isinstance(comp_unit, str):
                comp_labels.append(comp_unit)
                comp_aliases.append(f"{comp_unit}-component")

    # multi prop determination
    multi_prop_flag = None
    if isinstance(properties, list):
        if len(properties) > 1:
            multi_prop_flag = True
        elif len(properties) == 1:
            multi_prop_flag = False
    elif isinstance(properties,str):
        multi_prop_flag = False

    # system expansion flag
    systems_expansion_flag
    if len(systems_list) == 1:
        systems_expansion_flag = False
    elif len(systems_list) > 1 and systems_expansion is True:
        systems_expansion_flag = True
    else: systems_expansion_flag = False

    # expansion label determination
    expansion_label_set = ["rescale","properties","systems","single_comp"]
    expansion_label = None
    if rescale_flag is True:
        expansion_label = expansion_label_set[0]
    elif rescale_flag is False:
        if multi_prop_flag is True and systems_expansion_flag is False:
            expansion_label = expansion_label_set[2]
        elif multi_prop_flag is False and systems_expansion_flag is True:
            expansion_label = expansion_label_set[3]
```

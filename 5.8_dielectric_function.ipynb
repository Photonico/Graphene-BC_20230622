# Dielectric Function versus components

Dielectric function versus Photon energy in different components


```python
# Prework

from vmatplot.dielectric_function import plot_dielectric_function

import matplotlib.pyplot as plt

components = [{"xx":"in-plane"},{"zz":"out-of-plane"}]

layout = "hor"
boundary = 24
```

```python
# BC3

systems = [["", "5_Dielectric_function/A_BC3_PBE_K65", "blue"]]

plot_dielectric_function("BC₃", systems, components, layout, False, "eV", boundary)

plt.savefig("Figures_for_supplementary/S1.13a.pdf")
```

# Bandstructure with DoS

The band structure is demonstrated with Energy (in eV) versus K-point distance (in Å^2) with the Electronic density of state versus energy.


```python
# Prework

from vmatplot.bandstructure_plotting import plot_bsDoS

import matplotlib.pyplot as plt
```

```python
# Graphene-BC₃ (Hollow)

bsDoS_list = [["bands", "(PBE)", "3_Bandstructure/E_Graphene-BC3_PBE", "4_PDoS/E_Graphene-BC3_PBE","orbital"],
              ["bands", "(HSE06)", "3_Bandstructure/E_Graphene-BC3_HSE", "4_PDoS/E_Graphene-BC3_HSE", "orange"]]

plot_bsDoS("for Graphene-BC₃ (Hollow)", 6, 10, bsDoS_list, "upper right")

plt.savefig("Figures/3.1_bandstructure_Graphene-BC3.pdf")
```

```python
# Graphene-Borophene (Top)

bsDoS_list = [["bands", "(PBE)", "3_Bandstructure/F_Graphene-Borophene_PBE", "4_PDoS/F_Graphene-Borophene_PBE","orbital"],
              ["bands", "(HSE06)", "3_Bandstructure/F_Graphene-Borophene_HSE", "4_PDoS/F_Graphene-Borophene_HSE", "orange"]]

plot_bsDoS("for Graphene-Borphene (Top)", 6, 10, bsDoS_list, "upper right")

plt.savefig("Figures/3.2_bandstructure_Graphene-Borophene.pdf")
```

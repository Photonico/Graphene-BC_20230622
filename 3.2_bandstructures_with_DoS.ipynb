# Bandstructure with DoS

The band structure is demonstrated with Energy (in eV) versus K-point distance (in Å^2) with the Electronic density of state versus energy.


```python
# Prework

from vmatplot.bandstructure import plot_bsDoS

import matplotlib.pyplot as plt
```

```python
# Graphene-BC₃ (Hollow)

bsDoS_list = [["monocolor", "(GGA-PBE)", "3_Bandstructure/E_Graphene-BC3_hollow_PBE", "4_PDoS/E_Graphene-BC3_PBE","default"],
              ["monocolor", "(HSE06)", "3_Bandstructure/E_Graphene-BC3_hollow_HSE", "4_PDoS/E_Graphene-BC3_HSE", "orange"]]

plot_bsDoS("for Graphene-BC₃", 6, 10, bsDoS_list, None)

plt.savefig("Figures/3.21_bandstructure_Graphene-BC3.pdf")
```

```python
# Graphene-Borophene (Top)

bsDoS_list = [["monocolor", "(GGA-PBE)", "3_Bandstructure/F_Graphene-Borophene_top_PBE", "4_PDoS/F_Graphene-Borophene_PBE","default"],
              ["monocolor", "(HSE06)", "3_Bandstructure/F_Graphene-Borophene_top_HSE", "4_PDoS/F_Graphene-Borophene_HSE", "orange"]]

plot_bsDoS("for Graphene-Borphene", 6, 10, bsDoS_list, None)

plt.savefig("Figures/3.22_bandstructure_Graphene-Borophene.pdf")
```

```python
# Bilayer Graphene-B₄C₃ Top for Schottky Barrier

from vmatplot.bandstructure import plot_bandstructure

bandstructure_list = [["monocolor", "", "3_Bandstructure/G_Graphene-B4C3_top_HSE", "violet"]]

plot_bandstructure("for Graphene-B₄C₃", 3, bandstructure_list, None)

# plt.savefig("Figures/proj1.8b.pdf")
# plt.savefig("Figures/proj1.8b.svg")

fig = plt.gcf()
ax = plt.gca()
# Change figure size after creation
fig.set_size_inches(8, 6)

# Gamma position
x_gamma = ax.get_xticks()[-1]   # Rightmost Γ point
x_M = ax.get_xticks()[-2]       # M point
x_arrow = x_gamma - 0.05        # Place the arrow slightly to the left of Γ
x_text  = x_gamma - 0.05 * 4    # Place the text slightly to the left of Γ

# Energy values
E_F  = 0.0          # Fermi level
E_cb = 1.42         # Conduction band edge near the rightmost Γ point
E_vb = -0.87        # Valence band edge near the rightmost Γ point

ax.hlines(E_cb, x_gamma, x_M, colors="#1478E1", linestyles="--", zorder=10)
ax.hlines(E_vb, x_gamma, x_gamma - 0.04*2, colors="#EB731E", linestyles="--", zorder=10)

# Arrow of Phi_n
ax.annotate(
    "", xy=(x_arrow, E_cb), xytext=(x_arrow, E_F),
    arrowprops=dict(arrowstyle="<->", color="#1478E1", lw=1.5, mutation_scale=12)
)

# Arrow of Phi_p
ax.annotate(
    "", xy=(x_arrow, E_F), xytext=(x_arrow, E_vb),
    arrowprops=dict(arrowstyle="<->", color="#EB731E", lw=1.5, mutation_scale=12)
)

# text
ax.text(
    x_text, 0.5 * (E_cb + E_F), r"$\Phi_n$", color="#1478E1",
    fontsize=32, va="center", ha="left", clip_on=False
)

ax.text(
    x_text, 0.5 * (E_vb + E_F), r"$\Phi_p$", color="#EB731E",
    fontsize=32, va="center", ha="left", clip_on=False
)

plt.savefig("Figures/proj1.8b.pdf")
plt.show()
```

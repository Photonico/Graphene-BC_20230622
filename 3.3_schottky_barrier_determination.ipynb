# Schottky barrier determination

```python
# Prework

from vmatplot.bandstructure import plot_bandstructure, extract_high_sym_max_valence_intersections, extract_high_sym_min_conduction_intersections
```

```python
# Graphene-BC₃ GGA-PBE

project = "3_Bandstructure/E_Graphene-BC3_hollow_PBE"

bandstructure_list = [["bands", "", project, "blue"]]

plot_bandstructure("for Graphene-BC₃ with GGA-PBE", 6, bandstructure_list, "upper right")

max_valance = extract_high_sym_max_valence_intersections(project,"spin 1")

min_conduction = extract_high_sym_min_conduction_intersections(project,"spin 1")

print("For Graphene-BC₃ with GGA-PBE")

print(f"\n the minimum intersections of conduction bands: \n {min_conduction} \n")

print(f"\n the maximum intersections of valance bands: \n {max_valance} \n")
```

# Lattice

Free energy versus lattice constant

```python
## Graphene - BC₃ Top

## Read data
import numpy as np
from Store.lattice_extracting import read_lattice_distance
from scipy.interpolate import griddata

lattice, distance, free_energy = read_lattice_distance("0_Lattice_Twin_Distance/A1_Twin_Graphene-BC3_Top/lattice.dat")

## Vectorization of data
# Source data
lattice = np.array(lattice); distance = np.array(distance); free_energy = np.array(free_energy)
lattice_grid, distance_grid = np.meshgrid(np.unique(lattice), np.unique(distance), indexing="ij")
free_energy_grid = free_energy.reshape(len(np.unique(lattice)), -1)

# Extrema of source data
energy_min = np.min(free_energy)
energy_max = np.max(free_energy)
energy_range = energy_max - energy_min
energy_demo = energy_min + energy_range * 0.25
lattice_min = lattice [np.argmin(free_energy)]
distance_min = distance [np.argmin(free_energy)]

## Fitted data
# Prepare data for interpolation
points = np.c_[lattice.ravel(), distance.ravel()]
values = free_energy.ravel()

# Create a fine grid for interpolation
lattice_fine = np.linspace(lattice.min(), lattice.max(), 1024)
distance_fine = np.linspace(distance.min(), distance.max(), 1024)
lattice_grid_fine, distance_grid_fine = np.meshgrid(lattice_fine, distance_fine)

# Interpolate using the "cubic" method
free_energy_grid_fine = griddata((lattice, distance), free_energy, (lattice_grid_fine, distance_grid_fine), method="linear")

# Extrema of source data
energy_index_fitted_min = np.unravel_index(np.nanargmin(free_energy_grid_fine), free_energy_grid_fine.shape)
lattice_min_fitted = lattice_grid_fine[energy_index_fitted_min]
distance_min_fitted = distance_grid_fine[energy_index_fitted_min]
energy_min_fitted = free_energy_grid_fine[energy_index_fitted_min]

## Plot
import matplotlib.pyplot as plt

plt.figure(dpi=256, figsize=(10,6))
params = {"text.usetex": False, "font.family": "serif", "mathtext.fontset": "cm",
          "axes.titlesize": 18, "axes.labelsize": 14, "figure.facecolor": "w"}
plt.rcParams.update(params)
plt.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)
plt.title("Energy for Graphene - BC₃ Top"); plt.xlabel(r"Lattice constant (Å)"); plt.ylabel(r"Distance (Å)")

# Source data
cp = plt.pcolormesh(lattice_grid_fine, distance_grid_fine, free_energy_grid_fine, shading="auto", cmap="Blues_r", alpha = 0.75, vmax = energy_demo)
plt.colorbar(cp)

# Extreme of source data
plt.scatter(lattice_min, distance_min, s=60, linewidth=1.5, facecolors="none", edgecolors="#37AA3C", label="Extrema of source data", alpha=0.8, zorder=2)
plt.text(lattice_min, distance_min, f"\n Source extrema: \n ({lattice_min:.6f}, {distance_min:.6f}, {energy_min:.6f}) \n", ha="right", va="bottom")

# Extreme of fitted data
plt.scatter(lattice_min_fitted, distance_min_fitted, s=20, linewidth=1.5, facecolors="none", edgecolors="#F5820F", label="Extrema of fitted data", alpha=0.8, zorder=2)
plt.text(lattice_min_fitted, distance_min_fitted, f"\n Fitted extrema: \n ({lattice_min_fitted:.6f}, {distance_min_fitted:.6f}, {energy_min_fitted:.6f}) \n", ha="right", va="top")
```

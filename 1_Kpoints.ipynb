Free energy versus kpoints constant

```python
#### 1.1 Single layer - BC3

## Read data
import numpy as np
from Store.kpoints_extracting import read_kpoints

start = 3

kpoints, free_energy = read_kpoints("1_Kpoints/Single_BC3/Kpoints.dat")
kpoints = kpoints[start:]
free_energy = free_energy[start:]

## Source data specific features
free_energy_min = np.min(free_energy)
min_kpoints = kpoints[np.argmin(free_energy)]
free_energy_scale = np.max(free_energy)-np.min(free_energy)

## Plot
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

plt.figure(dpi=196, figsize=(10,6))
params = {"text.usetex": False, "font.family": "serif", "mathtext.fontset": "cm",
          "axes.titlesize": 18, "axes.labelsize": 14, "figure.facecolor": "w"}
plt.rcParams.update(params)
plt.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)
plt.title("Free energy versus kpoints for BC₃"); plt.xlabel(r"Lattice constant (Å)"); plt.ylabel(r"Energy (eV)")

plt.scatter(kpoints, free_energy, color="#1473D2", label="Source data", zorder=1)

formatter = ScalarFormatter(useMathText=True)
plt.gca().yaxis.set_major_formatter(formatter)

plt.ticklabel_format(style="sci", axis="y", scilimits=(-3,1))
plt.xticks(range(start+1, 20+1, 1))
plt.ylim(np.min(free_energy)-free_energy_scale*0.12, np.max(free_energy)+free_energy_scale*0.1)

plt.legend(loc="best")
```

# Extinction coefficient

The extinction coefficient, also known as the imaginary refractive index, indicates the attenuation of light in a system.

We define the extinction spectrum of a system as:

$$ K(\omega)=\sqrt{\left[\varepsilon_R^2(\omega)+\varepsilon_I^2(\omega)\right]^{\frac{1}{2}}-\frac{\varepsilon_I(\omega)}{2}} $$

or

$$ K(\omega)=\frac{\varepsilon_I(\omega)}{2\omega} $$

where:

$\omega$ is the given frequency;

$\varepsilon_R$ is real part dielectric function;

$\varepsilon_I$ is real part dielectric function.

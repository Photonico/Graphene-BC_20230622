# Reflectance

We define the reflectance of a system as:

$$ R(\omega)=\left|\frac{\sqrt{\varepsilon_R(\omega)+i\varepsilon_I(\omega)}-1}{\sqrt{\varepsilon_R(\omega)+i\varepsilon_I(\omega)}+1}\right|^2 $$

where:

$\omega$ is the given frequency;

$\varepsilon_R$ is real part dielectric function;

$\varepsilon_I$ is real part dielectric function.

```python
import os

def rename_folders_in_directory(directory="."):
    """Rename folders in the specified directory and its subdirectories."""
    for dirpath, dirnames, filenames in os.walk(directory, topdown=False):  # topdown=False to rename child folders first
        for dirname in dirnames:
            if "_var_" in dirname or "_dis_" in dirname:
                new_name = dirname.replace("_var_", "_a").replace("_dis_", "_d")
                old_folder_path = os.path.join(dirpath, dirname)
                new_folder_path = os.path.join(dirpath, new_name)
                os.rename(old_folder_path, new_folder_path)
                print(f"Renamed folder {old_folder_path} to {new_folder_path}")

# Call the function for the current directory
rename_folders_in_directory()
```

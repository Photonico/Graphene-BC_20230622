# Bandstructure

The band structure is demonstrated with Energy (in eV) versus K-point distance (in Å^2)


```python
# Monolayer BC₃

from vmatplot.bandstructure import plot_bandstructure

bandstructure_list = [["bands", "(PBE)", "3_Bandstructure/A_BC3_PBE", "blue"],
                      ["bands", "(HSE06)", "3_Bandstructure/A_BC3_HSE", "orange"]]

plot_bandstructure("for BC₃", 6, bandstructure_list, "upper right")
```

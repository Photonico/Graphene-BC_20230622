# Lattice

Total energy versus lattice constant


## Selected Results

### BC₃

* Most stable Lattice constant: 5.169 Angstrom
* Minimal free energy: -67.90772214 eV
* B-C bond: 1.563 Angstrom
* C-C bond: 1.421 Angstrom

### Graphene

* Most stable Lattice constant: 2.467 Angstrom
* Minimal free energy: -74.52125928 eV
* C-C bond: 1.424 Angstrom

### Borophene

* Most stable Lattice constant: 5.051 Angstrom
* Minimal free energy: -51.13821750 eV
* B-B bond: 1.689×4, 1.707×1, 1.672×2 Angstrom

### B₄C₃

* Most stable Lattice constant: 4.690 Angstrom
* Minimal free energy: -52.83670482 eV
* B-B bond: 1.689×1 Angstrom
* B-C bond: 1.548×1, 1.592×2 Angstrom


```python
#### Single layer - BC3

## Read data
import numpy as np
from Store_former.lattice_extracting import read_lattice, read_lattice_select
from Store_former.lattice_selecting import lattice_select

lattice, free_energy = read_lattice("0_Lattice_Single/A_BC3/lattice_free_energy.dat")
lattice_plot, free_energy_plot = read_lattice_select("0_Lattice_Single/A_BC3/lattice_free_energy.dat", 1000, 10)

## Select point
select_point = lattice_select("1_Lattice_Selected/A_BC3_a5.169")

print(select_point)

## Fit curve
# Fit a cubic polynomial to the data
p = np.polyfit(lattice, free_energy, 3)

# Generate a polynomial function from the fitted parameters
f = np.poly1d(p)

# Generate new x and y data using the polynomial function
lattice_fitted = np.linspace(min(lattice), max(lattice), num=1000, endpoint=True)
free_energy_fitted = f(lattice_fitted)

lattice_range = max(lattice)-min(lattice)

## Source data specific features
free_energy_min = np.min(free_energy)
lattice_min = lattice[np.argmin(free_energy)]
free_energy_scale = np.max(free_energy)-np.min(free_energy)

## Fitted data specific features
# Compute the derivative of the polynomial
f_prime = f.deriv()
# Find the roots of the derivative (where derivative equals zero)
roots = f_prime.r
fitted_lattice_extrema = roots[(roots >= min(lattice)) & (roots <= max(lattice))]
fitted_free_energy_extrema = f(fitted_lattice_extrema)

fitted_free_energy_min = np.min(fitted_free_energy_extrema)
fitted_lattice_min = fitted_lattice_extrema[np.argmin(fitted_free_energy_min)]

## Plot
import matplotlib.pyplot as plt

plt.figure(dpi=256, figsize=(10,6))
params = {"text.usetex": False, "font.family": "serif", "mathtext.fontset": "cm",
          "axes.titlesize": 18, "axes.labelsize": 14, "figure.facecolor": "w"}
plt.rcParams.update(params)
plt.tick_params(direction="in", which="both", top=True, right=True, bottom=True, left=True)
plt.title("Free energy versus lattice for BC₃"); plt.xlabel(r"Lattice constant (Å)"); plt.ylabel(r"Energy (eV)")

plt.scatter(lattice_plot, free_energy_plot, s=48, facecolors="#FFFFFF", edgecolors="#1473D2", label="Source data", zorder=2)
plt.scatter(lattice_min, free_energy_min, s=96, linewidth=1, facecolors="#FFFFFF", edgecolors="#1473D2", label="Extrema of source data", zorder=2)
plt.text(lattice_min+lattice_range*0.2, free_energy_min-free_energy_scale*0.04, f"Extrema of source data: {lattice_min:.6f}(Å), {free_energy_min:.6f}(eV)", color="#145AAA", ha="right", va="top")

plt.plot(lattice_fitted, free_energy_fitted, color="#1473D2", label="Fitted data", zorder=1)
plt.scatter(fitted_lattice_extrema, fitted_free_energy_extrema, s=48, linewidth=1, facecolors="#FFF", edgecolors="#1EB4FF", label="Extrema of fitted data", zorder=2)
plt.text(fitted_lattice_min+lattice_range*0.2, fitted_free_energy_min-free_energy_scale*0.04, f"\nExtrema of fitted data: {fitted_lattice_min:.6f}(Å), {fitted_free_energy_min:.6f}(eV)", color="#145AAA", ha="right", va="top")

plt.scatter(select_point[0], select_point[1], s=24, linewidth=1,  facecolors="#FFFFFF", edgecolors="#FA961E", label="Selected data", zorder=2)
plt.text(select_point[0]+lattice_range*0.2, select_point[1]-free_energy_scale*0.04 , f"\n\nSelected data: {select_point[0]:.6f}(Å), {select_point[1]:.6f}(eV)", ha="right", va="top")

plt.ylim(np.min(free_energy)-free_energy_scale*0.25, np.max(free_energy)+free_energy_scale*0.25)

plt.legend(loc="best")

```

# Absorption coefficient

We define the absorption coefficient of a system as:

$$ I(\omega) = \sqrt{2}\omega\left[\sqrt{\varepsilon_R^2(\omega)+\varepsilon_I^2(\omega)}-\varepsilon_R(\omega)\right] $$

where:

$\omega$ is the given frequency;

$\varepsilon_R$ is real part dielectric function;

$\varepsilon_I$ is real part dielectric function.

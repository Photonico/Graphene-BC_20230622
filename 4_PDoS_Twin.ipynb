## Element analysis result

The format of element is `number (start, end)`

where,

`number` refers to the number of this element in the current layer;

`start` refers to the start index of this element in the current layer;

`end` refers to the end index of this element in the current layer;

| Matter        | Direction | B in top layer    | C in top layer    | C in Graphene layer   |
| -             | -         | -                 | -                 | -                     |
| G - BC₃       | Hollow    | `2 (1, 2)`        | `6 (3, 8)`        | `8 (9, 16)`           |
| G - Borophene | Top       | `8 (1, 8)`        | -                 | `8 (9, 16)`           |
| G - B₄C₃      | Top       | `4 (1, 4)`        | `3 (5, 7)`        | `8 (8, 15)`           |

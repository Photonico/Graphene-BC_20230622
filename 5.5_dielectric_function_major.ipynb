# Dielectric Function versus components

Dielectric function versus Photon energy in different components


```python
# Prework

from vmatplot.dielectric_function import plot_dielectric_function

import matplotlib.pyplot as plt

components = [{"xx":"in-plane"},{"zz":"out-of-plane"}]

layout = "hor"
boundary = 24
```

```python
# BC₃

monolayers = [["for BC₃", "5_Dielectric_function/A_BC3_PBE_K65", "blue"]]

plot_dielectric_function("for BC₃", monolayers, components, layout, False, "eV", boundary)

plt.savefig("Figures/5.51_dielectric_BC3.pdf")
```

```python
# Borophene

monolayers = [["for Borophene", "5_Dielectric_function/B_Borophene_PBE_K65", "green"]]

plot_dielectric_function("for Borophene", monolayers, components, layout, False, "eV", boundary)

plt.savefig("Figures/5.52_dielectric_Borophene.pdf")
```

```python
# B₄C₃

monolayers = [["for B₄C₃", "5_Dielectric_function/C_B4C3_PBE_K65","violet"]]

plot_dielectric_function("for B₄C₃", monolayers, components, layout, False, "eV", boundary)

plt.savefig("Figures/5.53_dielectric_B4C3.pdf")
```

```python
# Graphene by GGA-PBE

monolayers = [["for Graphene", "5_Dielectric_function/D_Graphene_PBE_K129", "wine"]]

plot_dielectric_function("for Graphene", monolayers, components, layout, False, "eV", boundary)

plt.savefig("Figures/5.54_dielectric_Graphene.pdf")
```

```python
# Graphene-BC₃ (Hollow)

components = [{"xx":"in-plane"},{"zz":"out-of-plane"},"xy"]

systems = [["(PBE)", "5_Dielectric_function/E_Graphene-BC3_PBE_K65_Normal", "blue"],
           ["(HSE06 17 k-points)", "5_Dielectric_function/E_Graphene-BC3_HSE_K17", "orange"]]

plot_dielectric_function("for Graphene-BC₃", systems, components, layout, False, "eV", boundary)

plt.savefig("Figures/5.55_dielectric_Graphene-BC3.pdf")
```

```python
# Graphene-Borophene (Top)

components = [{"xx":"in-plane"},{"zz":"out-of-plane"}]

systems = [["(PBE)", "5_Dielectric_function/F_Graphene-Borophene_PBE_K65_Normal", "green"],
           ["(HSE06)", "5_Dielectric_function/F_Graphene-Borophene_HSE_K65_Accurate", "orange"]]

plot_dielectric_function("for Graphene-Borophene", systems, components, layout, False, "eV", boundary)

plt.savefig("Figures/5.56_dielectric_Graphene-Borophene.pdf")
```

# Dielectric Function versus algorithms and K-points

Dielectric function versus Photon energy in different kpoints and algorithms (GGA-PBE, HSE06)


```python
# Prework

from vmatplot.dielectric_function import plot_dielectric_function

import matplotlib.pyplot as plt
```

```python
# Graphene-Borophene with HSE06 versus accuracy configurations

dielectric_list = [["(PBE", "5_Dielectric_function/F_Graphene-Borophene_PBE_K65_Normal", "grey"],
                   ["(HSE06 17 k-points)", "5_Dielectric_function/F_Graphene-Borophene_HSE_K17", "green"],
                   ["(HSE06 Normal EDIFF = 1E-4)", "5_Dielectric_function/F_Graphene-Borophene_HSE_K65_Normal_EDIFF-4", "blue", "solid", 0.9, 3.2],
                   ["(HSE06 Normal EDIFF = 1E-5)", "5_Dielectric_function/F_Graphene-Borophene_HSE_K65_Normal_EDIFF-5", "violet", "solid", 1.0, 2.4],
                   ["(HSE06 Accurate EDIFF = 1E-6)", "5_Dielectric_function/F_Graphene-Borophene_HSE_K65_Accurate", "orange", "solid", 1.0, 0.8]]

plot_dielectric_function("for Graphene-Borophene in 65 k-points", dielectric_list, [{"xx":"in-plane"},{"zz":"out-of-plane"}], "hor", True, "eV", 24)

plt.savefig("Figures/5.21_dielectric_in_EDIFF.pdf")
```

```python
# Graphene-B₄C₃ (Top) with GGA-PBE versus accuracy configurations

dielectric_list = [["(25 k-points Normal)", "5_Dielectric_function/G_Graphene-B4C3_PBE_K25_Normal", "blue", "solid", 0.6, 4.4],
                   ["(25 k-points Exact)", "5_Dielectric_function/G_Graphene-B4C3_PBE_K25_Exact", "yellow"],
                  ]

plot_dielectric_function("for Graphene-B₄C₃ (Top) by PBE", dielectric_list, [{"xx":"in-plane"},{"zz":"out-of-plane"}], "hor", True, "eV", 24)

plt.savefig("Figures/5.22_dielectirc_in_precisions.pdf")
```

```python
# Graphene-B₄C₃ (Top) with GGA-PBE versus k-points

dielectric_list = [["(11 k-points)", "5_Dielectric_function/G_Graphene-B4C3_PBE_K11", "orange"],
                   ["(17 k-points)", "5_Dielectric_function/G_Graphene-B4C3_PBE_K17_N128", "blue"],
                   ["(25 k-points Exact)", "5_Dielectric_function/G_Graphene-B4C3_PBE_K25_Exact", "violet"],
                   ["(65 k-points Normal)", "5_Dielectric_function/G_Graphene-B4C3_PBE_K65_Normal", "Purple"],
                  ]

plot_dielectric_function("for Graphene-B₄C₃ (Top) by PBE", dielectric_list, [{"xx":"in-plane"},{"zz":"out-of-plane"}], "hor", True, "eV", 24)

plt.savefig("Figures/5.23_dielectirc_in_kpoints.pdf")
```
